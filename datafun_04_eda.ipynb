## 1) Imports 

```python
import pandas as pd
import seaborn as sns
import matplotlib 

# Axes object (basic plot type returned by Seaborn)
from matplotlib.axes import Axes
```

## 2) Load Data

```python
# Load the Iris dataset into pandas DataFrame
iris_df: pd.DataFrame = sns.load_dataset('iris')

# List column names
iris_df.columns

# Inspect first few rows of the DataFrame
iris_df.head()
```

## 3) Initial Data Inspection 

```python
# Specify the number of rows to display
iris_df.head(10)

# Inspect the shape of the DataFrame with shape attribute
# The shape is a tuple with count of rows and columns in the DataFrame
iris_df.shape

# Inspect the data types of the columns with dtypes attribute
# The data types are returned as a pandas Series
iris_df.dtypes

# Inspect the data types of the columns with info() method
iris_df.info()
```

## 4) Initial Descriptive Statistics 

```python
# Inspect summary statistics for numerical columns
iris_df.describe()
```

## 5) Initial Distribution for Numerical Columns

```python
# Inspect histogram by one numerical column
iris_df['sepal_length'].hist()

# Inspect histograms for ALL numerical columns
iris_df.hist()

# Show all plots
matplotlib.pyplot.show()
```

The petal length and width have either high values or low values, meaning the petals are typically either long and fat, or short and skinny comparatively. However the sepal is all around medium sized and proportional. 

## 6) Initial Data Transformation and Feature Engineering. 

```python
# Feature Engineering
# Renaming multiple columns
iris_df.rename(columns={'sepal_length': 'Sepal Length'}, inplace=True)

iris_df.rename(columns={'sepal_width': 'Sepal Width'}, inplace=True)

iris_df.rename(columns={'petal_length': 'Petal Length'}, inplace=True)

iris_df.rename(columns={'petal_width': 'Petal Width'}, inplace=True)

# Adding new columns
iris_df['Sepal Area'] = iris_df['Sepal Length'] * iris_df['Sepal Width']

iris_df['Petal Area'] = iris_df['Petal Length'] * iris_df['Petal Width']
```

## 7) Initial Visualizations 

```python
# Created a scatter plot to show the sepal's length versus its width
# The hue parameter is used to color the data points 
# by species (a categorical column)
scatter_plt = sns.scatterplot(iris_df, x='Sepal Length', y='Sepal Width', hue='species')

#setting the labels and title of the graph 
scatter_plt.set_xlabel("Sepal Length (mm)")
scatter_plt.set_ylabel("Sepal Width (mm)")
scatter_plt.set_title("Chart 1: Sepal Length vs Area")
# Show all plots
matplotlib.pyplot.show()
```

Initial Analysis: For the Setosa species, it's sepal is typically shorter and thinner than other species. The versicolor typically has a medium length sepal that stays on the thinner side. And the virginica's sepal is typically just as thin as the versicolor's it can be of the same length or longer. 

```python
# Created a scatter plot to show the sepal's length versus its width
# The hue parameter is used to color the data points 
# by species (a categorical column)
scatter_plts = sns.scatterplot(iris_df, x='Petal Length', y='Petal Width', hue='species')

#setting the labels and title of the graph 
scatter_plts.set_xlabel("Petal Length (mm)")
scatter_plts.set_ylabel("Petal Width (mm)")
scatter_plts.set_title("Chart 2: Petal Length vs Area")

# Show all plots
matplotlib.pyplot.show()
```

The setosa has a small and skinny petal. The versicolor has a medium sized petal. The virginica has the largest petal.

```python
# Created a histogram of the sepal area
# The hue parameter is used to color the data points 
# by species (a categorical column)
histplot_plt = sns.histplot(iris_df, x='Sepal Area', hue='species', multiple='dodge', kde=True)

#setting the labels and title of the graph 
histplot_plt.set_xlabel("Sepal Area (mm*2)")
histplot_plt.set_ylabel("Numer of Species with that Area")
histplot_plt.set_title("Chart 3: Sepal Areas Distribution")

# Show all plots
matplotlib.pyplot.show()
```

In the scatter plots above, the setosa seemed to be the species with the smallest sepal, but when we look at the shape distribution line in this histogram, we can see that that is not always the case. The setosa has less species with an area of 10.0 to around 14.5 than the versicolor. The setosa also has some organisms with a sepal area of 25, while the versicolor's sepal area doesn't go above 22.5. However, the histogram does confirm the inital observation that it has the largest sepal of the three species. 

```python
# Created a histogram of the sepal area
# The hue parameter is used to color the data points 
# by species (a categorical column)
histplot_plts = sns.histplot(iris_df, x='Petal Area', hue='species', multiple='dodge', kde=True)

#setting the labels and title of the graph 
histplot_plts.set_xlabel("Petal Area (mm*2)")
histplot_plts.set_ylabel("Numer of Species with that Area")
histplot_plts.set_title("Chart 3: Petal Areas Distribution")

# Show all plots
matplotlib.pyplot.show()
```

The histogram of the species' petal area does match our previous observations. The setosa has the smallest petal, the cersicolor has a medium petal, and the viginica has the largest petal.
